#Data Cleaning


```python
#Check if nulls in some columns cause No intrnet service
for i,j,h in zip( data['Internet Service'] , data['Internet Type'],data['Avg Monthly GB Download']):
  if i=='No':
    print(i,j,h)
```

```python
# Replace null values with suitable values
def replace_nulls(col__name,val) :
  data[col__name]=data[col__name].replace(np.nan,val)

replace_nulls('Internet Type','no internet service')
replace_nulls('Avg Monthly GB Download',0)
replace_nulls('Online Security','no internet service')
replace_nulls('Online Backup','no internet service')
replace_nulls('Device Protection Plan','no internet service')
replace_nulls('Premium Tech Support','no internet service')
replace_nulls('Streaming TV','no internet service')
replace_nulls('Streaming Movies','no internet service')
replace_nulls('Streaming Music','no internet service')
replace_nulls('Unlimited Data','no internet service')
```

```python
#fill null values in column 'Avg Monthly Long Distance Charges' with median
#data['Avg Monthly Long Distance Charges']=data['Avg Monthly Long Distance Charges'].fillna(data['Avg Monthly Long Distance Charges'].mean()) # mean and median are the same value 25.69
data['Avg Monthly Long Distance Charges']=data['Avg Monthly Long Distance Charges'].fillna(data['Avg Monthly Long Distance Charges'].sort_values(ascending=True).median())
data['Avg Monthly Long Distance Charges']
```

```python
#Check if null values in column 'Multiple Lines' dependes on column 'Phone Service'
for i,j in zip( data['Phone Service'] ,data['Multiple Lines']):
  if i=='No':
    print(i,j)
```

```python
# Replace null values in column 'Multiple Lines' with txt 'no phone service'
def replace_nulls(col__name,val) :
  data[col__name]=data[col__name].replace(np.nan,val)

replace_nulls('Multiple Lines','no phone service')
```

```python
#Drop useless columns
data=data.drop(['Latitude',"Longitude","Churn Reason","Churn Category","Offer"],axis="columns")
```

# Analysis and Visualization

```python
# What is Customer status distribution
pivot1=pd.pivot_table(data,index="Customer Status",aggfunc='size')
plt.figure(figsize=(6,6))
plt.pie(pivot1,labels=pivot1.index,colors=['#bca88e','#e4b4b4','#9da993'],autopct='%1.1f%%', startangle=140);
plt.title('Customer Status Distribution')
plt.show()
```

```python
#Top 10 Cities with the Highest Churn Rates
pivot=pd.pivot_table(data,index='City',columns='Customer Status',aggfunc='size')
pivot['Total_Customers'] = pivot.sum(axis=1)
top_10_cities = pivot.sort_values(by='Total_Customers', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.bar(top_10_cities.index, top_10_cities['Total_Customers'],color='#bca88e')
plt.bar(top_10_cities.index, top_10_cities['Churned'])
plt.bar(top_10_cities.index, top_10_cities['Stayed'],color='gray')
plt.bar(top_10_cities.index, top_10_cities['Joined'])
plt.xlabel('City')
plt.ylabel('Number of Customers')
plt.title('Top 10 Cities with the Highest Churn Rates')
plt.legend(['Total Customers', 'Churned', 'Stayed', 'Joined'])
plt.xticks(rotation=90);
```

```python
#Average age per customer segments
#data['Customer ID'].is_unique
Custumer_status_By_Age=pd.pivot_table(data,index='Customer Status',values='Age',aggfunc='mean')
plt.barh(Custumer_status_By_Age.index,Custumer_status_By_Age['Age'],color='gray')
plt.xlabel('Age')
plt.ylabel('Customer Status')
plt.title('Average age and customers gender distribution per customer segments')
```

```python
#create a new column refers to customer tenure
def classify_tenure(tenure):
  if tenure < 5:
        return 'New Customer'
  elif 5 <= tenure <= 10:
        return 'short-Term'
  elif 10 <= tenure <= 20:
        return 'Medium-Term'
  else:
        return 'Long-Term'
data['Customer Tenure Segments']=data['Tenure in Months'].apply(classify_tenure)
data['Customer Tenure Segments']
```

```python
#churn rate distribution based on customers tenure classification
Customer_Tenure_Segments=pd.pivot_table(data,index='Customer Tenure Segments', columns="Customer Status",aggfunc='size')

plt.figure(figsize=(10,6))
plt.bar(Customer_Tenure_Segments.index,Customer_Tenure_Segments['Churned'],color='#bca88e')
plt.xlabel('Customer Tenure Segments')
plt.ylabel('Number of Customers')
plt.title('Churn Rate Distribution Based on Customers Tenure Classification')
plt.legend(['Churned'])
```

```python
# Churned Customers based on internet service and phone service status
internet_service=pd.pivot_table(data,index='Internet Service',columns='Customer Status',aggfunc='size')
phone_service=pd.pivot_table(data,index='Phone Service',columns='Customer Status',aggfunc='size')

fig,ax= plt.subplots(1,2, figsize=(10,4))
ax[0].bar(internet_service.index,internet_service['Churned'],color='gray')
ax[0].set_title('Chrned Customers by Internet Srvice')
ax[0].set_xlabel('Internet Service')
ax[0].set_ylabel('Number of Churned Customers')
ax[1].bar(phone_service.index,phone_service['Churned'],color='brown')
ax[1].set_title('Chrned Customers by Internet Srvice')
ax[1].set_xlabel('Phone Service')

```
